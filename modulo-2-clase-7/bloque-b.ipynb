```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

cargamos el archivo

```python
cvd_df = pd.read_csv("cvd_cleaned.csv")
```

```python
pd.set_option('display.max_columns', None)
cvd_df.head()
```

```python
cvd_df.info()
```

## Gr√°fico de Barras

Luego procedemos a preparar nuestro dataset para graficar:

```python
hacen_ejercicio = (
    (cvd_df[cvd_df["Exercise"] == "Yes"])[["General_Health", "Exercise"]]
    .groupby(by=["General_Health"])
    .max()
    .reset_index()
)
hacen_ejercicio.head()
```

Y luego procedemos a graficar (en eje x dejamos las categor√≠as y en eje y dejamos la cuenta de personas que realizan ejercicio):

```python
sns.barplot(data=hacen_ejercicio, x="General_Health", y="Exercise")
plt.show()
```

## Boxplot

```python
sns.boxplot(data=cvd_df, x="BMI", y="General_Health")
plt.show()
```

Introduciremos algunas opciones para mejorar nuestros gr√°ficos:

```python
figure = plt.figure(figsize=(10,5))
sns.boxplot(data=cvd_df, x="BMI", y="General_Health", hue='General_Health')
plt.xlabel("√çndice de masa corporal")
plt.ylabel("Estado de salud")
plt.title("Boxplot para las distribuciones de √≠ndice de masa corporal seg√∫n estado de salud", fontweight='bold')
plt.show()
```
