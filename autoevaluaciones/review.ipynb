```python
def leer_autoevaluacion(nombre_archivo):
    name = nombre_archivo.split('_')[-1].replace('.xlsx', '')
    wb = load_workbook(nombre_archivo)
    ws = wb['AUTOEVALUACIÓN']
    estudiante = ' '.join([i[0].value for i in ws['B4:C4']])
    if estudiante == 'NOMBRE : ':
        estudiante = ws[''].value
    estudiante = estudiante.replace('NOMBRE : ', '')
    
    evaluaciones = [
        'trabajo_colaborativo',
        'participacion_actividades',
        'compromisos_tiempo_forma_contenido',
        'dificultades_superar',
        'satisfecho_trabajo'
    ]
    notas = list(map(lambda l: [x.value for x in l][0], ws['D9:D13']))
    autoevaluacion_dict = {
        'estudiante': [name,],
    }
    for i in range(len(evaluaciones)):
        autoevaluacion_dict[evaluaciones[i]] = [notas[i],]
    autoevaluacion_df = pd.DataFrame(autoevaluacion_dict)
    return autoevaluacion_df
```

```python
leer_autoevaluacion('data/Autoevaluacion_Coevaluacion_AlvaroPerez.xlsx')
```

Leer todas las autoevaluaciones

```python
df = pd.DataFrame()
for archivo in glob('data/*.xlsx'):
    if not archivo.startswith('data/~$'):
        df = pd.concat([df, leer_autoevaluacion(archivo)])
df.reset_index(drop=True)
df
```

### Coevaluación

Prueba con un archivo

```python
wb = load_workbook('data/Autoevaluacion_Coevaluacion_AlvaroPerez.xlsx')
# obtener el nombre de las hojas del archivo
print(wb.sheetnames)
ws = wb['AUTOEVALUACIÓN']
ws_coevaluacion = wb['COEVALUACIÓN']
```

```python
co_evaluaciones = [
    'Participó activamente en el trabajo colaborativo',
    'Participó con entusiasmo en las actividades del curso',
    'Entregó sus compromisos en tiempo, forma y contenido dentro del trabajo grupal',
    'Se esforzó en superar sus dificultades',
    'Me siento satisfecho(a) con el trabajo que el(ella) ha realizado'
]

estudiante = ' '.join([i[0].value for i in ws['B4:C4']])
estudiante = estudiante.replace('NOMBRE :', '')
companeros = list(map(lambda l: [x.value for x in l], ws_coevaluacion['H9:N9']))[0]
notas = list(map(lambda l: [x.value for x in l], ws_coevaluacion['H10:N14']))
notas_dict = {
    'companeros': companeros,
    'evaluador':[estudiante] * len(companeros)
}
for i in range(len(co_evaluaciones)):
    notas_dict[co_evaluaciones[i]] = notas[i]
notas_df = pd.DataFrame(notas_dict)
notas_df.head()
```

```python
def leer_coevaluacion(nombre_archivo):
    wb = load_workbook(nombre_archivo)
    ws_coevaluacion = wb['COEVALUACIÓN']
    ws = wb['AUTOEVALUACIÓN']
    estudiante = ' '.join([i[0].value for i in ws['B4:C4']])
    estudiante = estudiante.replace('NOMBRE :', '')
    companeros = list(map(lambda l: [x.value for x in l], ws_coevaluacion['H9:N9']))[0]
    notas = list(map(lambda l: [x.value for x in l], ws_coevaluacion['H10:N14']))
    notas_dict = {
        'companeros': companeros,
        'evaluador':[estudiante] * len(companeros)
    }
    for i in range(len(co_evaluaciones)):
        notas_dict[co_evaluaciones[i]] = notas[i]
    notas_df = pd.DataFrame(notas_dict)
    return notas_df
```

```python
leer_coevaluacion('data/Autoevaluacion_Coevaluacion_AlvaroPerez.xlsx')
```

Leer todas las coevaluaciones

```python
df_coevaluacion = pd.DataFrame()
for archivo in glob('data/*.xlsx'):
    if not archivo.startswith('data/~$'):
        df_coevaluacion = pd.concat([df_coevaluacion, leer_coevaluacion(archivo)])
df_coevaluacion.reset_index(drop=True)
df_coevaluacion
```
